```python
import random #Standard libaries.
import numpy as np
import time
import copy

from scipy.signal import convolve2d, correlate2d  #Importing functions to perform efficent 2D convolution and cross-correlation.
from scipy.linalg import block_diag #Importing a function to quickly construct block diagonal matrices.
from skimage.measure import block_reduce #Importing a function to perform efficient max-pooling.

import pickle  #Libraries for importing data sets.
import gzip

def load_data():
    """Return the MNIST data as a tuple containing the training data,
    the validation data, and the test data.
    The ``training_data`` is returned as a tuple with two entries.
    The first entry contains the actual training images.  This is a
    numpy ndarray with 50,000 entries.  Each entry is, in turn, a
    numpy ndarray with 784 values, representing the 28 * 28 = 784
    pixels in a single MNIST image.
    The second entry in the ``training_data`` tuple is a numpy ndarray
    containing 50,000 entries.  Those entries are just the digit
    values (0...9) for the corresponding images contained in the first
    entry of the tuple.
    The ``validation_data`` and ``test_data`` are similar, except
    each contains only 10,000 images.
    This is a nice data format, but for use in neural networks it's
    helpful to modify the format of the ``training_data`` a little.
    That's done in the wrapper function ``load_data_wrapper()``, see
    below.
    """
    f = gzip.open('..\Coding\mnist.pkl.gz', 'rb')
    training_data, validation_data, test_data = pickle.load(f, encoding='bytes')
    f.close()
    return (training_data, validation_data, test_data)

def load_data_wrapper():
    """Return a tuple containing ``(training_data, validation_data,
    test_data)``. Based on ``load_data``, but the format is more
    convenient for use in our implementation of neural networks.
    In particular, ``training_data`` is a list containing 50,000
    2-tuples ``(x, y)``.  ``x`` is a 784-dimensional numpy.ndarray
    containing the input image.  ``y`` is a 10-dimensional
    numpy.ndarray representing the unit vector corresponding to the
    correct digit for ``x``.
    ``validation_data`` and ``test_data`` are lists containing 10,000
    2-tuples ``(x, y)``.  In each case, ``x`` is a 784-dimensional
    numpy.ndarry containing the input image, and ``y`` is the
    corresponding classification, i.e., the digit values (integers)
    corresponding to ``x``.
    Obviously, this means we're using slightly different formats for
    the training data and the validation / test data.  These formats
    turn out to be the most convenient for use in our neural network
    code."""
    tr_d, va_d, te_d = load_data()
    training_inputs = [np.reshape(x, (784, 1)) for x in tr_d[0]]
    training_results = [vectorized_result(y) for y in tr_d[1]]
    training_data = list(zip(training_inputs, training_results))
    validation_inputs = [np.reshape(x, (784, 1)) for x in va_d[0]]
    validation_data = list(zip(validation_inputs, va_d[1]))
    test_inputs = [np.reshape(x, (784, 1)) for x in te_d[0]]
    test_data = list(zip(test_inputs, te_d[1]))
    return (training_data, validation_data, test_data)


def vectorized_result(j):
    """Return a 10-dimensional unit vector with a 1.0 in the jth
    position and zeroes elsewhere.  This is used to convert a digit
    (0...9) into a corresponding desired output from the neural
    network."""
    e = np.zeros((10, 1))
    e[j] = 1.0
    return e

class Network(object):
    def __init__(self, layers, activation_function = sigmoid, padding = False, dropout_frequency = 0, batch_normalisation = False):  #Initialises the attributes (variables) of each network object we create.
        global convolution_time
        self.num_layers = len(layers)
        self.layers = layers
        self.padding = padding  #Here we can choose whether or not to turn on zero-padding in any convolutional steps. 
        self.activation_function = activation_function
        if activation_function == relu: self.activation_function_prime = relu_prime
        if activation_function == sigmoid: self.activation_function_prime = sigmoid_prime
        #Our weights are now initially randomly generated using an improved normal distribution (note that we use the same distribution as before for the biases since it makes little difference).
        #The new initial weight value distribution is on a node-by-node basis with mean 0 and s.d. given by 1/sqrt(no. of input weights to node).
        #This means that our weighted input 'z' into the node is much more likely to be a small number (around -1 to 1 usually) which avoids large activation values and reduces node saturation.
        self.weights = [None for layer in layers[:-1]]  #Initialising storage for our weights and biases.
        self.biases = [None for layer in layers[1:]]
        if batch_normalisation == True:
            self.gammas = [None for layer in layers[:-2]] #Initialising storage for our gamma  and beta variables in case we are using batch normalisation.
            self.betas = [None for layer in layers[:-2]] #Initialising storage for our gamma  and beta variables in case we are using batch normalisation.
            self.epsilon = 1e-4 #Value of epsilon added to the variance in the division part of the batch normalisation step.
            self.global_mean = 0 #Setting default values of the global mean and variance used for feedforward in evaluation and gameplay (these are updated during learning).
            self.global_variance = 1 
        if type(dropout_frequency) == list:
            self.dropout_frequency = dropout_frequency[0]
            self.input_layer_dropout = dropout_frequency[1]
        else: 
            self.dropout_frequency = dropout_frequency
            self.input_layer_dropout = False
        self.batch_normalisation = batch_normalisation
        #Here we don't allow our final layer to be convolutional as it needs to have a node for every choice. Typically this isn't a strict requirement for a network but I'm enforcing it here to make things slightly simpler
        if type(layers[-1]) == list:
            print("Final layer must not be a convolutional layer.")
            return
        for i in range(self.num_layers-1):
            #We can describe a convolutional layer in our network by entering a 6 component list entry in our array 'layers'.
            #These components should be: the number of filters in the layer, the x and y dimensions of our array of input activations, the size of each filter (for simplicity we only consider square filters),
            #the size of the pooling filter and the stride at which the pooling filter moves across our convolution outputs.
            if type(layers[i]) == list:   #Checking if the current layer is a convolutional layer.
                if len(layers[i]) < 4:   #Here we make sure that the correct data is given to describe our convolutional layer.
                    print("Convolutional layer is missing required arguments: num_filters, layer_x_size, layer_y_size, filter_size. Optional arguments also include pooling_size, pooling_stride (with default values 1).")
                    return
                if len(layers[i]) > 6:
                    print("Too many input arguments for convolutional layer. Allowed input arguments are: num_filters, layer_x_size, layer_y_size, filter_size, pooling_size, pooling_stride.")
                    return
                while len(layers[i]) < 6:  #If we don't specify a pooling size or a pooling stride then we set these to have value 1 by default.
                    layers[i].append(1)
                [num_filters, layer_x_size, layer_y_size, filter_size, pooling_size, pooling_stride] = layers[i]  #Unpacking our convolutional layer information.
                layer_size = layer_x_size*layer_y_size
                #In the convolutional layer we need a bias for every node in the layer and we also need filter_size weights per filter as well as a next layer bias for every filter.
                self.biases[i] = np.array(np.random.randn(num_filters, 1),dtype=np.float32) #Here np.random.randn(x, 1) makes a x-by-1 array of random numbers.
                self.weights[i] = np.array(np.random.randn(num_filters, filter_size*filter_size)/np.sqrt(layer_size),dtype=np.float32)  #Here we normalise by our filter size in each filter.
                #We need to ensure that the output of our convolutional layer fits the input of the next layer. Note that whether or not we use padding will effect the number of outputs of our convolutional layer.
                layer_dimensions, filter_dimensions = [layer_x_size, layer_y_size], [filter_size, filter_size]
                #Determining the size we will need for the layer following our convolutional layer. 
                conv_layer_output_size = convolution_output_shape(layer_dimensions, filter_dimensions, padding)
                if pooling_size > 1:  #If we are adding pooling after our convolution then the size of the output will also depend on the pooling size and stride.
                    for j in [0, 1]:
                        conv_layer_output_size[j] = 1 + np.ceil((conv_layer_output_size[j]-pooling_size)/pooling_stride)
                conv_layer_output_size = int(num_filters*conv_layer_output_size[0]*conv_layer_output_size[1])
                #Checking to see whether the current layer is compatible with the output of the previous convolutional layer.
                if type(layers[i+1]) == list: next_layer_input_size = layers[i+1][1]*layers[i+1][2]
                else: next_layer_input_size = layers[i+1]
                if conv_layer_output_size != next_layer_input_size:
                    print("A layer after a convolutional layer doesn't match the size of the expected convolutional layer output: ",next_layer_input_size,"!=",conv_layer_output_size)
                    return
            else:
                if type(layers[i+1]) == list: next_layer_size = layers[i+1][1]*layers[i+1][2]  #Determining the size needed for inputs to our layer if it's a convolutional layer.                    
                else: next_layer_size = layers[i+1]  #Otherwise we have a fully connected layer so its size is simply prescribed by the value of layers[i-1].
                self.biases[i] = np.array(np.random.randn(next_layer_size, 1),dtype=np.float32) #Note that we don't define biases for fully connected first layer nodes.
                self.weights[i] = np.array(np.random.randn(next_layer_size, layers[i])/np.sqrt(layers[i]),dtype=np.float32) 
                if batch_normalisation == True and i != self.num_layers-2: 
                    self.gammas[i] = np.array(np.random.randn(layers[i], 1),dtype=np.float32) #Randomly initialising gamma values for all layers but our last two layers.
                    self.betas[i] = np.array(np.random.randn(layers[i], 1),dtype=np.float32) #Randomly initialising gamma values for all layers but our last two layers.
        self.experiences = []

#This accepts an input data vector 'a' and for each layer in our network we evaluate a new value of 'a'
#using the tuple (b,w) := (vector of node biases, array of previous layer node weights feeding in to each 
#of our nodes). This means thaft at each step 'a' functions as our current vector of node activations.
    def feedforward(self, a):
        global convolution_time
        layer = 0
        if self.batch_normalisation == True:
            gamma = self.gammas[0]
            beta = self.betas[0]
            if self.global_mean is 0: normalised_a = a #Accounting for when we are feeding forward without having done any training (so we don't have any global means/variances).
            else: normalised_a = (a-self.global_mean[0])/np.sqrt(self.global_variance[0]+self.epsilon)
            a = gamma*normalised_a+beta             
        for b, w in zip(self.biases, self.weights):   #The arrays of both the bias vectors and weight matrices have length num_layers-1 since there are no first layer biases and no final layer weights.
            if type(self.layers[layer]) == list:   #If the current layer is described by a list then it is a convolutional layer so we perform a convolution to get the activation.
                [num_filters, layer_x_size, layer_y_size, filter_size, pooling_size, pooling_stride] = self.layers[layer]  #Unpacking our convolutional layer information.
                 #We convolve (or technically cross-correlate) each of our filters with 2D array of activations fed into our convolutional layer by our previous layer to get a weighted input for each filter.
                reshaped_a = np.reshape(a, (layer_y_size, layer_x_size))
                a = np.array((),dtype=np.float32)
                for j in range(num_filters):  #The weights for each filter are stored in separate vectors contained in the entry of 'weights' corresponding to the convolutional layer.
                    convolution_time_start = time.time()
                    convolution_filter = np.reshape(w[j], (filter_size, filter_size))
                    if self.padding == True: z = correlate2d(reshaped_a, convolution_filter, 'full') + b[j]  #If we have padding enabled then we obtain an outputs that include the cases where our filter isn't entirely overlapping with our 2D array of nodes, i.e. when it's hanging off the edge.
                    if self.padding == False: z = correlate2d(reshaped_a, convolution_filter, 'valid') + b[j] #Otherwise we only obtain outputs from calculations made when our filter exactly fits onto our 2D array of nodes.
                    convolution_time = convolution_time + time.time() - convolution_time_start
                    if pooling_size > 1: z = max_pooling(z, pooling_size, pooling_stride) #If a pooling filter size is prescribed we then apply max-pooling to the output of the convolution of our filter with our input activations.
                    a = np.append(a, self.activation_function(z))   #Finally we apply our ReLU activation function to the pooled output and append this to our array of activations. 
                a = np.reshape(a, (np.size(a), 1))
            else:
                #If we are using batch normalisation then we need to normalise our weighted inputs for each layer and scale them using gamma and beta. 
                #During the testing phase we normalise using the mean and standard deviation from the whole training set (rather than from a mini-batch).
                if self.batch_normalisation == True and layer < self.num_layers-3:
                    z = np.matmul(w, a)
                    gamma = self.gammas[layer+1]
                    beta = self.betas[layer+1]
                    if self.global_mean is 0: normalised_z = z #Accounting for when we are feeding forward without having done any training (so we don't have any global means/variances).
                    else: normalised_z = (z-self.global_mean[layer+1])/np.sqrt(self.global_variance[layer+1]+self.epsilon)
                    z = gamma*normalised_z+beta 
                    # print("ff: using gamma, beta in layer",layer+1)
                else: 
                    z = np.matmul(w, a)+b
                    # print("using bias for feedforward in layer ",layer)
                if layer == self.num_layers-2 and self.activation_function == relu:   #This holds if we are looking at weights and biases in the final layer.
                    # print("not applying activation function in layer ",layer+1)
                    a = z   #We don't apply our activation function to our final layer weighted inputs so that our outputs take the form of values (as in our tabular case)
                else:
                    # print("applying activation function in layer ",layer+1)
                    a = self.activation_function(z)  #For all but our final layer we apply our ReLU activation function which introduces non-linearity into our network.                             
            layer = layer + 1  #Incrementing our layer counter as we feedforward through our network
#Here a is a vector since np.matmul(w, a) and b are vectors and the activation function is performed elementwise.
        return a

    def SGD(self, training_data, epochs, mini_batch_size, eta,
            test_data = None, display_weights = False, lmbda = 0):
        """Train the neural network using mini-batch stochastic
        gradient descent.  The ``training_data`` is a list of tuples
        ``(x, y)`` representing the training inputs and the desired
        outputs.  The other non-optional parameters are
        self-explanatory.  If ``test_data`` is provided then the
        network will be evaluated against the test data after each
        epoch, and partial progress printed out.  This is useful for
        tracking progress, but slows things down substantially."""
        global convolution_time, pooling_time, index_time
        if test_data: n_test = len(test_data)
        n = len(training_data)

        for j in range(epochs):
            convolution_time, pooling_time, index_time = 0, 0, 0
            start_time = time.time()
            random.shuffle(training_data)
            mini_batches = [
                training_data[k:k+mini_batch_size]
                for k in range(0, n, mini_batch_size)]
            global_mean, global_variance = 0, 0
            for mini_batch in mini_batches:
                mb_means, mb_variances = self.update_mini_batch(mini_batch, eta, lmbda, n) #If we aren't using batch normalisation then means and variances will return as None.
                if self.batch_normalisation == True:
                    global_mean = np.add(global_mean, mb_means)
                    global_variance = np.add(global_variance, mb_variances)
            #If we are using batch normalisation then we need to calculate the mean and variance of our entire training set for use in the testing phase later.
            #We calculate the population mean/variance using averages from all of the mini-batch means/averages calculated during training.
            self.global_mean = global_mean/len(mini_batches)
            self.global_variance = mini_batch_size/(mini_batch_size-1)*global_variance/len(mini_batches)
            if test_data:
                print("Epoch {0}: {1} / {2}".format(
                    j+1, self.evaluate(test_data), n_test))
            else:
                print("Epoch {0} complete".format(j+1))
            total_time = time.time() - start_time
            print("Time taken =",total_time)
            for layer in self.layers:
                if type(layer) == list:
                    print("Convolution time =",convolution_time)
                    if layer[5] != 1:
                        print("Pooling time =",pooling_time)
                        print("Index time =",index_time)
                        print("Other time =",total_time-convolution_time-pooling_time)
                        break
                    print("Other time =",total_time-convolution_time)
                    break
            for layer in range(self.num_layers):
                if type(self.layers[layer]) == list:
                    total_similarity = get_similarity(self.weights[layer]) 
                    print("Total filter similarity =",total_similarity)
            if display_weights == True: 
                self.display_weights(True)
            print("")
                
            
    def update_mini_batch(self, mini_batch, eta, lmbda, n):
        """Update the network's weights and biases by applying
        gradient descent using backpropagation to a single mini batch.
        The ``mini_batch`` is a list of tuples ``(x, y)``, and ``eta``
        is the learning rate."""
        grad_b = [np.zeros(b.shape) for b in self.biases]
        grad_w = [np.zeros(w.shape) for w in self.weights]
        # if self.batch_normalisation == False: #Standard backpropagation without batch normalisation.
        # if len(mini_batch)%2==1:
        #     for x, y in mini_batch:
        #         delta_grad_b, delta_grad_w = self.backprop(x, y)
        #         grad_b = [gb+dgb for gb, dgb in zip(grad_b, delta_grad_b)]
        #         grad_w = [gw+dgw for gw, dgw in zip(grad_w, delta_grad_w)]
        # # if self.batch_normalisation == True: #Backpropagation with batch normalisation.
        if self.batch_normalisation == False: 
            grad_b, grad_w = self.batch_backprop(mini_batch)
            self.weights = [(1-eta*(lmbda/n))*w-(eta/len(mini_batch))*gw for w, gw in zip(self.weights, grad_w)]
            self.biases = [b-(eta/len(mini_batch))*gb for b, gb in zip(self.biases, grad_b)]  
            return None, None
        if self.batch_normalisation == True: 
            grad_gamma, grad_beta, grad_b, grad_w, mb_means, mb_variances = self.batch_backprop(mini_batch)
            # print("grad_gamma =",grad_gamma)
            # print("grad_beta =",grad_beta)
            # print("grad_w =",grad_w)
            self.biases = [b-(eta/len(mini_batch))*gb for b, gb in zip(self.biases, grad_b)]  
            self.weights = [(1-eta*(lmbda/n))*w-(eta/len(mini_batch))*gw for w, gw in zip(self.weights, grad_w)]
            self.gammas = [gamma-(eta/len(mini_batch))*ggamma for gamma, ggamma in zip(self.gammas, grad_gamma)]
            self.betas = [beta-(eta/len(mini_batch))*gbeta for beta, gbeta in zip(self.betas, grad_beta)]
            return mb_means, mb_variances

        #For convolutional layers we encourage filters to learn different weights by punishing them for having similar weights (and also normalising the vector of weights for each filter).
        # for layer in range(self.num_layers-1):
        #     if type(self.layers[layer]) == list and lmbda != 0:   
        #         sim_gradient = np.array(get_similarity_gradient(self.weights[layer]))
        #         self.weights[layer] = self.weights[layer]-eta*(nabla_w[layer]+lmbda*sim_gradient)/len(mini_batch)
        #         self.weights[layer] = self.weights[layer]/np.linalg.norm(self.weights[layer])  #Here we normalise the weights (this acts similarly to regularisation and keeps the similarity punshing process stable).
        #     else: self.weights[layer] = self.weights[layer]-eta*nabla_w[layer]/len(mini_batch)
        # self.biases = [b-(eta/len(mini_batch))*nb for b, nb in zip(self.biases, nabla_b)]

    #Backpropagates through our network taking in one mini batch tuple input (x,y) at a time to calculate the gradient
#vectors wrt our weights and biases for the cost function according to the four backpropagation equations.
    def backprop(self, x, y): 
        global convolution_time
        grad_b = [np.array(np.zeros(np.shape(b)),dtype=np.float32) for b in self.biases] #Initialises the gradient vector for our biases.
        grad_w = [np.array(np.zeros(np.shape(w)),dtype=np.float32) for w in self.weights] #Initialises the gradient vector for our weights.
#Here we feedforward to calculate our weighted inputs z^l and our activations a^l := activation_function(z) for each layer l,
#then using these and our backpropagation equations we can calculate the final layer error vector delta^L and the
#rest of our layer error vectors delta^l for all l.
        activation = x #This sets our mini batch inputs as our first layer activation vector.
        activations = [activation] #This creates a matrix to store each of our layer activations.
        zs = [] #This creates an array to store all of our weighted input vectors z (or z^l).
        max_pooling_indices = [] #This creates an array to keep track of any maximal indices used in any max pooling steps we perform (we need these later for backpropagtion).
        if self.dropout_frequency > 0: #If we are using dropout then we create a 'mask' for each hidden layer which tell us which neurons to drop in each layer (this will vary per training example).
            masks = [] #Creating an array to hold the mask for each layer.
            if self.input_layer_dropout == True: #This only runs if we want to use dropout on our input layer. 
                input_layer_mask = np.random.binomial(1, 1-self.dropout_frequency, (self.layers[0], 1))/(1-self.dropout_frequency)
                activation = activation * input_layer_mask 
            for l in range(self.num_layers-3): #In this case we don't perform dropout on the first (input) layer, final (output) layer or the penultimate layer so we only create (up to) num_layers-3 masks.
                if type(self.layers[l+1]) == list: masks.append(None) #We won't perform dropout on any convolutional layers. 
                else: masks.append(np.random.binomial(1, 1-self.dropout_frequency, (self.layers[l+1], 1))/(1-self.dropout_frequency)) #We divide by the dropout frequency since we are using 'inverted dropout'.       
        layer = 0
        for b, w in zip(self.biases, self.weights):   #The arrays of both the bias vectors and weight matrices have length num_layers-1 since there are no first layer biases and no final layer weights.
            max_z_indices = []
            if type(self.layers[layer]) == list:   #If the current layer is described by a list then it is a convolutional layer so we perform a convolution to get the activation.
                [num_filters, layer_x_size, layer_y_size, filter_size, pooling_size, pooling_stride] = self.layers[layer]  #Unpacking our convolutional layer information.
                 #We convolve (or technically cross-correlate) each of our filters with 2D array of activations fed into our convolutional layer by our previous layer to get a weighted input for each filter.
                reshaped_activation = np.reshape(activation, (layer_y_size, layer_x_size))
                z = np.array((),dtype=np.float32)
                activation = np.array((),dtype=np.float32)
                for f in range(num_filters):  #The weights for each filter are stored in separate vectors contained in the entry of 'weights' corresponding to the convolutional layer.
                    convolution_time_start = time.time()
                    convolution_filter = np.reshape(w[f], (filter_size, filter_size))  
                    if self.padding == True: z_f = correlate2d(reshaped_activation, convolution_filter, 'full') + b[f]  #If we have padding enabled then we obtain an outputs that include the cases where our filter isn't entirely overlapping with our 2D array of nodes, i.e. when it's hanging off the edge.
                    if self.padding == False: z_f = correlate2d(reshaped_activation, convolution_filter, 'valid') + b[f] #Otherwise we only obtain outputs from calculations made when our filter exactly fits onto our 2D array of nodes.
                    convolution_time += time.time() - convolution_time_start
                    if pooling_size > 1: 
                        z_f, max_z_f_indices = max_pooling(z_f, pooling_size, pooling_stride, True) #If a pooling filter size is prescribed we then apply max-pooling to the output of the convolution of our filter with our input activations.
                        max_z_indices.append(max_z_f_indices)
                    z = np.append(z, z_f)  #Appending the weighted input from this filter to our overall weighted input.
                    activation = np.append(activation, self.activation_function(z_f))   #We then apply our activation function to the output obtained by convolving our filter with our input activations and then append this to the array of activations.
                z = np.reshape(z, (np.size(z), 1))  #Reshaping our overall z and activation into the format that works with our backpropagation algorithm.
                activation = np.reshape(activation, (np.size(activation), 1))
            else:
                z = np.matmul(w, activation)+b   #Calculating our weighted input (for a non-convolutional layer).
                if layer == self.num_layers-2 and self.activation_function == relu:  activation = z   #We don't apply our activation function to our final layer weighted inputs so that our outputs take the form of values (as in our tabular case)
                else: activation = self.activation_function(z)  #For all but our final layer we apply our activation function which introduces non-linearity into our network.
            activations.append(activation)
            #We choose whether to apply dropout to our first layer and we never apply dropout to our output layer or the penultimate layer. Also the saved activations for backpropagation aren't the dropout activations since the backpropagation dropout is achieved by applying our mask to the delta passed back to the dropout layer.
            if self.dropout_frequency > 0 and layer < self.num_layers-3: 
                if masks[layer] is not None: 
                    activation = activation * masks[layer] #Applying dropout to hidden layers on the forward pass.
            zs.append(z)                    
            max_pooling_indices.append(max_z_indices)
            layer = layer + 1
        if self.activation_function == relu: delta = self.cost_derivative(activations[-1], y)
        else: delta = self.cost_derivative(activations[-1], y) * self.activation_function_prime(zs[-1])#Calculates our final layer error vector for the cross-entropy cost function using BP1.
        #Note that here we are multiplying two vectors lengthways to create a matrix grad_w[-1] with entries dC/dw_jk^L at the jth row and kth column.
        #Here we backpropagate one layer at a time to calculate the error delta for each layer using BP2 and then we calculate our cost gradient vectors grad_b and grad_w for each layer using BP3 and BP4 respectively.
        #First we use our last layer delta to calculate the next delta and pass it on: since the last layer is forced to not be convolutional this is just our standard delta calculation.
        for l in range(1, self.num_layers): 
            if l != self.num_layers-1: z = zs[-l-1] #Unless we are on the first layer we will need our previous layer weighted input at each step.
            if type(self.layers[-l-1]) == list:  #Convolutional layer calculations.
                #Using the current layer delta to calculate grad_b and grad_w for a convolutional layer.
                [num_filters, layer_x_size, layer_y_size, filter_size, pooling_size, pooling_stride] = self.layers[-l-1]  #Unpacking the data for our convolutional layer.
                #Calculating the delta to be passed down to the next layer; note that as our current layer is a convolutional layer the calculations are more complicated than before.
                layer_dimensions, filter_dimensions = [layer_x_size, layer_y_size], [filter_size, filter_size]
                delta_f_sizes = convolution_output_shape(layer_dimensions, filter_dimensions, self.padding)  #Calculating the dimension that each delta_f array should have when reshaped.
                if pooling_size > 1:   #If pooling was used then the dimensions in delta_f_sizes will be smaller than as calculated above.
                    padding_dimensions = pooling_padding_needed(delta_f_sizes, (pooling_size, pooling_size), pooling_stride)
                    delta_f_sizes = (int((delta_f_sizes[1]+padding_dimensions[1]-pooling_size+pooling_stride)/pooling_stride), int((delta_f_sizes[0]+padding_dimensions[0]-pooling_size+pooling_stride)/pooling_stride))
                delta_f_size = delta_f_sizes[0]*delta_f_sizes[1]
                delta_fs = []  #Creating an array to store the delta arrays for each filter.
                reshaped_activation = np.reshape(activations[-l-1], (layer_y_size, layer_x_size))
                for f in range(num_filters):  #For convolutional layers we build up our gradients using the deltas for each filter (i.e. the sections of the overall delta).
                    convolution_filter = np.reshape(self.weights[-l][f], (filter_size, filter_size))  
                    delta_f = np.reshape(delta[f*delta_f_size:(f+1)*delta_f_size], (delta_f_sizes[1], delta_f_sizes[0]))
                    #If pooling has been used then we essentially need to undo our pooling step by increasing the size of delta_f and inserting zeros at the entries of the non-maximal elements.
                    if pooling_size > 1:   #Pooling layer delta_f calculations; here we only have non-zero entries of delta_f corresponding to the activations which were maximal in the max pooling step.
                        convolution_time_start = time.time()
                        if self.padding == True: z_f = correlate2d(reshaped_activation, convolution_filter, 'full') + self.biases[-l][f]  #If we have padding enabled then we obtain an outputs that include the cases where our filter isn't entirely overlapping with our 2D array of nodes, i.e. when it's hanging off the edge.
                        if self.padding == False: z_f = correlate2d(reshaped_activation, convolution_filter, 'valid') + self.biases[-l][f] #Otherwise we only obtain outputs from calculations made when our filter exactly fits onto our 2D array of nodes.
                        convolution_time += time.time() - convolution_time_start
                        unpooled_delta_f = np.zeros(np.shape(z_f))   #Initialising an array the size of our convolutional layer's convolution output.
                        max_indices = max_pooling_indices[-l][f]
                        k = 0  #Counter to move through the max_indices array.
                        for i in range(delta_f_sizes[1]):    #Looping over y values.
                            for j in range(delta_f_sizes[0]):   #Looping over x values.
                                if max_indices[k] is not None:
                                    unpooled_delta_f[max_indices[k][0],max_indices[k][1]] = delta_f[i,j]
                                k = k + 1
                        delta_f = unpooled_delta_f
                    grad_b[-l][f] = sum(delta_f.flatten())   #Calculating grad_b for the bias associated with filter f.
                    convolution_time_start = time.time()
                    # if self.padding == True: grad_w[-l][f] = correlate2d(reshaped_activation, delta_f, 'full').flatten()   #Calculating grad_w for the weights associated with filter f.
                    grad_w[-l][f] = correlate2d(reshaped_activation, delta_f, 'valid').flatten()   #Calculating grad_w for the weights associated with filter f.
                    convolution_time += time.time() - convolution_time_start
                    # #We always use padding in our backpropagation steps so that no information is lost.
                    # We obtain a contribution to the delta to be passed to our next layer by convolving the individual deltas for each filter with that filter.
                    if l != self.num_layers-1:
                        delta_f_sizes = convolution_output_shape(np.shape(convolution_filter), delta_f_sizes, True)
                        #We use convolution to calculate the next delta to be passed down for each filter.
                        convolution_time_start = time.time()
                        delta_f = convolve2d(convolution_filter, delta_f, 'full')*np.reshape(self.activation_function_prime(z), delta_f_sizes)
                        convolution_time += time.time() - convolution_time_start
                        delta_fs.append(delta_f)
                    #The overall delta to be passed down to the next layer is then the elementwise sum of all of the delta_fs. Note that for our last layer we can skip this step as there is no further layer to pass on a delta to.
                if l != self.num_layers-1:   
                    delta = np.zeros(np.shape(delta_fs[0]))
                    for df in delta_fs:
                        delta = delta + df
                    delta = np.reshape(delta, (np.size(delta), 1))                
            else:   #Fully connected layer calculations.
                #Using the current layer delta to calculate grad_b and grad_w for a fully connected layer.
                grad_b[-l] = delta
                grad_w[-l] = np.matmul(delta, activations[-l-1].transpose())   #Again this creates a matrix since we are multiplying our vectors lengthways.
                #We then calculate the value of delta to be passed down from this layer to the next. Note that for our last layer we can skip this step as there is no further layer to pass on a delta to.
                if l != self.num_layers-1: delta = np.matmul(self.weights[-l].transpose(), delta)*self.activation_function_prime(z)  
            #If dropout was used on a layer in our forward pass then we apply dropout to that layer's delta which makes the entries of delta corresponding to any dropped nodes equal to zero.
            if self.dropout_frequency > 0:
                if l != self.num_layers-1 and l != 1: 
                    if masks[-l+1] is not None:
                        delta = delta * masks[-l+1] #Applying dropout to hidden layer deltas on the backward pass.
                if self.input_layer_dropout == True and l == self.num_layers: delta = delta * input_layer_mask #This only runs for backpropagation for our input layer if we chose our input layer to be affected by dropout.
        return (grad_b, grad_w) #Finally we return a tuple of our cost gradient vectors wrt our biases and matrices wrt our weights for our whole network.
    

    
    
    def batch_backprop(self, mini_batch): #Old
        global convolution_time
        m = len(mini_batch)
        if self.batch_normalisation == False:
            grad_b = [np.array(np.zeros(np.shape(b)),dtype=np.float32) for b in self.biases] #Initialises the gradient vector for our biases.
            grad_w = [np.array(np.zeros(np.shape(w)),dtype=np.float32) for w in self.weights] #Initialises the gradient vector for our weights.
        mb_activation, mb_y = [], []
        for x, y in mini_batch:
            if len(mb_activation) == 0: mb_activation = x
            else: mb_activation = np.concatenate((mb_activation, x))
            if len(mb_y) == 0: mb_y = y
            else: mb_y = np.concatenate((mb_y, y))
        # print("first layer activations before =",mb_activation)
        # print(sum(mb_activation))
        final_mb_zs = []
        if self.batch_normalisation == True:
            grad_b = [np.array(np.zeros(np.shape(b)),dtype=np.float32) for b in self.biases] #Initialises the gradient vector for our biases.
            grad_w = [np.array(np.zeros(np.shape(w)),dtype=np.float32) for w in self.weights] #Initialises the gradient vector for our weights.
            grad_gamma = [np.array(np.zeros(np.shape(gamma)),dtype=np.float32) for gamma in self.gammas] #Initialises the gradient vector for our gammas.
            grad_beta = [np.array(np.zeros(np.shape(beta)),dtype=np.float32) for beta in self.betas] #Initialises the gradient vector for our betas.
            normalised_mb_zs = [] #Array to store the normalised and fully batch normalised mini batch weighted inputs after performing batch normalisation.
            layer_means = [np.array(np.zeros(np.shape(gamma)),dtype=np.float32) for gamma in self.gammas] #Array to store activation means for each layer.
            layer_variances = [np.array(np.zeros(np.shape(gamma)),dtype=np.float32) for gamma in self.gammas] #Array to store activation variances for each layer.
            mb_activation, normalised_mb_z, means, variances = batch_normalise(mb_activation, self.epsilon, self.gammas[0], self.betas[0], m)
            normalised_mb_zs.append(normalised_mb_z)
            # print("using gamma, beta in layer 0")
            # print("normalised_mb_zs input =",normalised_mb_z)
            # print("means = ",means)
            layer_means[0] = means  #We save the mean and variance from this layer's batch normalisation for use in backpropagation later.
            layer_variances[0] = variances
        final_mb_zs.append(mb_activation)
        # print("first layer activations after =",mb_activation)
        # print(sum(mb_activation))
        mb_activations = [mb_activation] #This creates a matrix to store each of our layer activations.
        mb_zs = [] #Array to store all of our weighted input vectors z (or z^l).
        max_pooling_indices = [] #This creates an array to keep track of any maximal indices used in any max pooling steps we perform (we need these later for backpropagtion).
        layer = 0
        for b, w in zip(self.biases, self.weights):   #The arrays of both the bias vectors and weight matrices have length num_layers-1 since there are no first layer biases and no final layer weights.
            max_z_indices = []
            if type(self.layers[layer]) == list:   #If the current layer is described by a list then it is a convolutional layer so we perform a convolution to get the activation.
                [num_filters, layer_x_size, layer_y_size, filter_size, pooling_size, pooling_stride] = self.layers[layer]  #Unpacking our convolutional layer information.
                 #We convolve (or technically cross-correlate) each of our filters with 2D array of activations fed into our convolutional layer by our previous layer to get a weighted input for each filter.
                reshaped_activation = np.reshape(activation, (layer_y_size, layer_x_size))
                z = np.array((),dtype=np.float32)
                activation = np.array((),dtype=np.float32)
                for f in range(num_filters):  #The weights for each filter are stored in separate vectors contained in the entry of 'weights' corresponding to the convolutional layer.
                    convolution_time_start = time.time()
                    convolution_filter = np.reshape(w[f], (filter_size, filter_size))  
                    if self.padding == True: z_f = correlate2d(reshaped_activation, convolution_filter, 'full') + b[f]  #If we have padding enabled then we obtain an outputs that include the cases where our filter isn't entirely overlapping with our 2D array of nodes, i.e. when it's hanging off the edge.
                    if self.padding == False: z_f = correlate2d(reshaped_activation, convolution_filter, 'valid') + b[f] #Otherwise we only obtain outputs from calculations made when our filter exactly fits onto our 2D array of nodes.
                    convolution_time += time.time() - convolution_time_start
                    if pooling_size > 1: 
                        z_f, max_z_f_indices = max_pooling(z_f, pooling_size, pooling_stride, True) #If a pooling filter size is prescribed we then apply max-pooling to the output of the convolution of our filter with our input activations.
                        max_z_indices.append(max_z_f_indices)
                    z = np.append(z, z_f)  #Appending the weighted input from this filter to our overall weighted input.
                    activation = np.append(activation, self.activation_function(z_f))   #We then apply our activation function to the output obtained by convolving our filter with our input activations and then append this to the array of activations.
                z = np.reshape(z, (np.size(z), 1))  #Reshaping our overall z and activation into the format that works with our backpropagation algorithm.
                activation = np.reshape(activation, (np.size(activation), 1))
            else:
                layer_size = np.shape(w)[1]
                mb_z = []
                if self.batch_normalisation == True and layer < self.num_layers-3:
                    # print("batch normalised feedforward on layer ",layer)
                    for i in range(m):
                        z = np.matmul(w, mb_activation[i*layer_size:(i+1)*layer_size])   #If we are using batch normalisation then we don't need to use biases (they cancel out in the batch normalisation step).
                        if len(mb_z) == 0: mb_z = z
                        else: mb_z = np.concatenate((mb_z, z))
                    mb_zs.append(mb_z)
                    gamma = self.gammas[layer+1]
                    beta = self.betas[layer+1]
                    # print("using gamma, beta in layer",layer+1)
                    mb_z, normalised_mb_z, means, variances = batch_normalise(mb_z, self.epsilon, gamma, beta, m)
                    layer_means[layer+1] = np.reshape(means, (np.size(means),1))  #We save the mean and variance from this layer's batch normalisation for use in backpropagation later.
                    layer_variances[layer+1] = np.reshape(variances, (np.size(variances),1))
                    normalised_mb_zs.append(normalised_mb_z)
                    final_mb_zs.append(mb_z)
                else: 
                    # print("using bias in layer of size ",layer_size)
                    for i in range(m):
                        z = np.matmul(w, mb_activation[i*layer_size:(i+1)*layer_size])+b #Calculating our weighted input (for a non-convolutional layer).
                        if len(mb_z) == 0: mb_z = z
                        else: mb_z = np.concatenate((mb_z, z))
                    mb_zs.append(mb_z) 
                    final_mb_zs.append(mb_z)
                if layer == self.num_layers-2 and self.activation_function == relu:  mb_activation = mb_z   #We don't apply our activation function to our final layer weighted inputs so that our outputs take the form of values (as in our tabular case)
                else: mb_activation = self.activation_function(mb_z)  #For all but our final layer we apply our activation function which introduces non-linearity into our network.
            mb_activations.append(mb_activation)
            max_pooling_indices.append(max_z_indices)
            layer = layer + 1
        # print("mb_zs: ",len(mb_zs), np.shape(mb_zs[0]))
        # print("final_mb_zs: ",len(final_mb_zs), np.shape(final_mb_zs[0]))
        if self.activation_function == relu: 
            mb_delta = self.cost_derivative(mb_activations[-1], mb_y)
            print("first mb_delta shape = ",np.shape(mb_delta))
        else: mb_delta = self.cost_derivative(mb_activations[-1], mb_y) * self.activation_function_prime(mb_zs[-1])#Calculates our final layer error vector for the cross-entropy cost function using BP1.
        #Note that here we are multiplying two vectors lengthways to create a matrix grad_w[-1] with entries dC/dw_jk^L at the jth row and kth column.
        #Here we backpropagate one layer at a time to calculate the error delta for each layer using BP2 and then we calculate our cost gradient vectors grad_b and grad_w for each layer using BP3 and BP4 respectively.
        #First we use our last layer delta to calculate the next delta and pass it on: since the last layer is forced to not be convolutional this is just our standard delta calculation.
        # print("mb_delta =",np.shape(mb_delta))
        # print("layer means: ",len(layer_means), np.shape(layer_means[0]))
        # print("layer variances: ",len(layer_variances), np.shape(layer_variances[0]))
        # print("mb_zs: ",len(mb_zs), np.shape(mb_zs[0]))
        # print("normalised_mb_zs: ",len(normalised_mb_zs), np.shape(normalised_mb_zs[0]))
        for l in range(1, self.num_layers): 
            if l != self.num_layers-1: mb_z = mb_zs[-l-1] #Unless we are on the first layer we will need our previous layer weighted input at each step.
            if type(self.layers[-l-1]) == list:  #Convolutional layer calculations.
                #Using the current layer delta to calculate grad_b and grad_w for a convolutional layer.
                [num_filters, layer_x_size, layer_y_size, filter_size, pooling_size, pooling_stride] = self.layers[-l-1]  #Unpacking the data for our convolutional layer.
                #Calculating the delta to be passed down to the next layer; note that as our current layer is a convolutional layer the calculations are more complicated than before.
                layer_dimensions, filter_dimensions = [layer_x_size, layer_y_size], [filter_size, filter_size]
                delta_f_sizes = convolution_output_shape(layer_dimensions, filter_dimensions, self.padding)  #Calculating the dimension that each delta_f array should have when reshaped.
                if pooling_size > 1:   #If pooling was used then the dimensions in delta_f_sizes will be smaller than as calculated above.
                    padding_dimensions = pooling_padding_needed(delta_f_sizes, (pooling_size, pooling_size), pooling_stride)
                    delta_f_sizes = (int((delta_f_sizes[1]+padding_dimensions[1]-pooling_size+pooling_stride)/pooling_stride), int((delta_f_sizes[0]+padding_dimensions[0]-pooling_size+pooling_stride)/pooling_stride))
                delta_f_size = delta_f_sizes[0]*delta_f_sizes[1]
                delta_fs = []  #Creating an array to store the delta arrays for each filter.
                reshaped_activation = np.reshape(activations[-l-1], (layer_y_size, layer_x_size))
                for f in range(num_filters):  #For convolutional layers we build up our gradients using the deltas for each filter (i.e. the sections of the overall delta).
                    convolution_filter = np.reshape(self.weights[-l][f], (filter_size, filter_size))  
                    delta_f = np.reshape(delta[f*delta_f_size:(f+1)*delta_f_size], (delta_f_sizes[1], delta_f_sizes[0]))
                    #If pooling has been used then we essentially need to undo our pooling step by increasing the size of delta_f and inserting zeros at the entries of the non-maximal elements.
                    if pooling_size > 1:   #Pooling layer delta_f calculations; here we only have non-zero entries of delta_f corresponding to the activations which were maximal in the max pooling step.
                        convolution_time_start = time.time()
                        if self.padding == True: z_f = correlate2d(reshaped_activation, convolution_filter, 'full') + self.biases[-l][f]  #If we have padding enabled then we obtain an outputs that include the cases where our filter isn't entirely overlapping with our 2D array of nodes, i.e. when it's hanging off the edge.
                        if self.padding == False: z_f = correlate2d(reshaped_activation, convolution_filter, 'valid') + self.biases[-l][f] #Otherwise we only obtain outputs from calculations made when our filter exactly fits onto our 2D array of nodes.
                        convolution_time += time.time() - convolution_time_start
                        unpooled_delta_f = np.zeros(np.shape(z_f))   #Initialising an array the size of our convolutional layer's convolution output.
                        max_indices = max_pooling_indices[-l][f]
                        k = 0  #Counter to move through the max_indices array.
                        for i in range(delta_f_sizes[1]):    #Looping over y values.
                            for j in range(delta_f_sizes[0]):   #Looping over x values.
                                if max_indices[k] is not None:
                                    unpooled_delta_f[max_indices[k][0],max_indices[k][1]] = delta_f[i,j]
                                k = k + 1
                        delta_f = unpooled_delta_f
                    grad_b[-l][f] = sum(delta_f.flatten())   #Calculating grad_b for the bias associated with filter f.
                    convolution_time_start = time.time()
                    # if self.padding == True: grad_w[-l][f] = correlate2d(reshaped_activation, delta_f, 'full').flatten()   #Calculating grad_w for the weights associated with filter f.
                    grad_w[-l][f] = correlate2d(reshaped_activation, delta_f, 'valid').flatten()   #Calculating grad_w for the weights associated with filter f.
                    convolution_time += time.time() - convolution_time_start
                    # #We always use padding in our backpropagation steps so that no information is lost.
                    # We obtain a contribution to the delta to be passed to our next layer by convolving the individual deltas for each filter with that filter.
                    if l != self.num_layers-1:
                        delta_f_sizes = convolution_output_shape(np.shape(convolution_filter), delta_f_sizes, True)
                        #We use convolution to calculate the next delta to be passed down for each filter.
                        convolution_time_start = time.time()
                        delta_f = convolve2d(convolution_filter, delta_f, 'full')*np.reshape(self.activation_function_prime(z), delta_f_sizes)
                        convolution_time += time.time() - convolution_time_start
                        delta_fs.append(delta_f)
                    #The overall delta to be passed down to the next layer is then the elementwise sum of all of the delta_fs. Note that for our last layer we can skip this step as there is no further layer to pass on a delta to.
                if l != self.num_layers-1:   
                    delta = np.zeros(np.shape(delta_fs[0]))
                    for df in delta_fs:
                        delta = delta + df
                    delta = np.reshape(delta, (np.size(delta), 1))                
            else:   #Fully connected layer calculations.
                #Using the current layer delta to calculate grad_b and grad_w for a fully connected layer.
                # print("mb_zs[-1] =",np.shape(mb_zs[-1]))
                # print("normalised_mb_zs[-1] =",np.shape(normalised_mb_zs[-1]))
                # print("mb_activations[-1] =",np.shape(mb_activations[-1]))
                # print("layer means[-1] =",np.shape(layer_means[-1]))
                # print("layer variances[-1] =",np.shape(layer_variances[-1]))
                w_shape = np.shape(self.weights[-l])
                # print("w_shape =",w_shape)
                layer_size = w_shape[1]
                next_layer_size = w_shape[0]
                #Calculating the small changes required for this layer's weights (these are needed regardless of whether we use batch normalisation or not).
                mb_grad_w = np.zeros((m, w_shape[0], w_shape[1]))
                # print("mb_grad_w =",np.shape(mb_grad_w))
                # if l == self.num_layers-1: print("testing for mb_activations[-l-1]: ",mb_activations[-l-1])
                # print("////////////// calculating grad w")
                mb_grad_y = mb_delta
                if self.batch_normalisation == True:
                    #If we are using batch normalisation then there are many intermediate calculations we need to do before calculating the previous layer's value of mb_delta.    
                    if l > 2: #For all layers but the final two we recieve an mb_delta that represents dC/dy, hence to recover our actual mb_delta=dC/dx we need to use the batch normalisation backprop equations.
                        # print("calculating fancy delta for l =",l)
                        # print("mb_grad_y =",np.shape(mb_grad_y))
                        #We then calculate mb_delta=dC/dx to be passed down to the next layer.
                        mb_z = mb_zs[-l]
                        final_mb_z = final_mb_zs[-l+2]
                        mb_means = np.tile(layer_means[-l+2], (m,1))
                        # print("layer means =",layer_means)
                        # print("layer_means[-l+1] =",np.shape(layer_means[-l+2]))                       
                        # print("layer_variances[-l+1] =",np.shape(layer_variances[-l+2]))
                        mb_sqrt_variances = np.tile(np.sqrt(layer_variances[-l+2] + self.epsilon), (m,1))
                        mb_gammas = np.tile(self.gammas[-l+2], (m,1))
                        # print("mb_gammas =",np.shape(mb_gammas))
                        # print("mb_delta =",np.shape(mb_delta))
                        grad_normalised_mb_z = mb_gammas*mb_grad_y
                        # print("grad_normalised_mb_z =", np.shape(grad_normalised_mb_z))
                        # print("mb_z =",np.shape(mb_z))
                        # print("mb_means =",np.shape(mb_means))
                        # print("mb_z-mb_means =",np.shape(mb_z-mb_means))
                        # print("mb_sqrt_variances =",np.shape(mb_sqrt_variances))
                        grad_variances = -0.5*np.reshape(np.sum(np.reshape(grad_normalised_mb_z*(mb_z-mb_means)*mb_sqrt_variances**(-3), (m, next_layer_size)), 0), (next_layer_size, 1))
                        # print("grad_variances =",np.shape(grad_variances))
                        grad_means = np.sum(np.reshape(-grad_normalised_mb_z/mb_sqrt_variances, (m, next_layer_size)), 0, keepdims=True).T + grad_variances*np.sum(-2*(mb_z-mb_means), 0, keepdims=True)/m
                        # print("grad_means =",np.shape(grad_means))
                        mb_grad_means = np.tile(grad_means, (m,1))
                        mb_grad_variances = np.tile(grad_variances, (m,1))
                        grad_mb_z = grad_normalised_mb_z/mb_sqrt_variances + 2*mb_grad_variances*(mb_z-mb_means)/m + mb_grad_means/m
                        # print("grad_mb_z =",np.shape(grad_mb_z))
                        mb_delta = grad_mb_z#*self.activation_function_prime(final_mb_z)
                #We can then use the correct value of mb_delta to calculate the changes to our weights and biases.
                # print("calculating grad_w and grad_b for l =",l)
                for i in range(m):
                    print("mb_delta shape = ",np.shape(mb_delta))
                    print("mb_delta[...] shape = ", np.shape(mb_delta[i*next_layer_size:(i+1)*next_layer_size]))
                    print("next layer size =",next_layer_size)
                    print("layer size =",layer_size)
                    mb_grad_w[i] = np.matmul(mb_delta[i*next_layer_size:(i+1)*next_layer_size], np.reshape(mb_activations[-l-1].T[0][i*layer_size:(i+1)*layer_size], (1, layer_size)))
                reshaped_mb_grad_w = np.reshape(mb_grad_w, (m, next_layer_size, layer_size))
                grad_w[-l] = np.sum(reshaped_mb_grad_w, 0)
                reshaped_mb_delta = np.reshape(mb_delta, (m, next_layer_size))
                grad_b[-l] = np.sum(reshaped_mb_delta, 0)
                grad_b[-l] = np.reshape(grad_b[-l], (np.size(grad_b[-l]), 1))
                # print("grad_w, grad_b calculated for layer l =",l)
                # print("back prop delta calculations for layer of size ",layer_size)
                # print("old mb_delta = ",np.shape(mb_delta))
                #We then calculate the value of mb_delta to be passed down from this layer to the next. Note that for our last layer we can skip this step as there is no further layer to pass on a delta to.
                mb_z = final_mb_zs[-l-1]
                new_mb_delta = []
                for i in range(m): 
                    # print("self.weights[-l].T =",np.shape(self.weights[-l].T))
                    # print("mb_delta[i*next_layer_size:(i+1)*next_layer_size]) =",np.shape(mb_delta[i*next_layer_size:(i+1)*next_layer_size]))
                    # print("mb_z[i*layer_size:(i+1)*layer_size] =",np.shape(mb_z[i*layer_size:(i+1)*layer_size]))
                    if l == self.num_layers-1:
                        # print("applying no activation for l =",l)
                        delta = np.matmul(self.weights[-l].T, mb_delta[i*next_layer_size:(i+1)*next_layer_size])#*self.activation_function_prime(mb_z[i*layer_size:(i+1)*layer_size])    #Calculating our weighted input (for a non-convolutional layer).
                    else:
                        # print("applying activation for l =",l)
                        delta = np.matmul(self.weights[-l].T, mb_delta[i*next_layer_size:(i+1)*next_layer_size])*self.activation_function_prime(mb_z[i*layer_size:(i+1)*layer_size])    #Calculating our weighted input (for a non-convolutional layer).
                    if len(new_mb_delta) == 0: new_mb_delta = delta
                    else: new_mb_delta = np.concatenate((new_mb_delta, delta))
                mb_delta = new_mb_delta
                # print("=================")
                # print("new mb_delta = ",np.shape(mb_delta))
                # print("l = ",l)
                #Calculating the small changes required for this layer's beta and gamma values using the value of mb_delta passed back from the next layer.
                # print("========== calculating grad beta/gamma")
                if self.batch_normalisation == True and l > 2:
                    # print("grad_gamma = ",np.shape(grad_gamma[-l+2]),"for l =",l)
                    # print("calculating grad_beta and grad_gamma for layer l=",l)
                    # print("mb_grad_y shape =",np.shape(mb_grad_y))
                    reshaped_mb_grad_y = np.reshape(mb_grad_y, (m, next_layer_size))
                    grad_beta[-l+2] = np.sum(reshaped_mb_grad_y.T, 1, keepdims=True)
                    # print("normalised_mb_zs[-l+1] =",np.shape(normalised_mb_zs[-l+2]))
                    reshaped_normalised_mb_z = np.reshape(normalised_mb_zs[-l+2], (m, next_layer_size))
                    # print("reshaped_mb_delta =",np.shape(reshaped_mb_delta))
                    grad_gamma[-l+2] = np.sum((reshaped_mb_grad_y*reshaped_normalised_mb_z).T, 1, keepdims=True)
                    # print("grad_gamma = ",np.shape(grad_gamma[-l+2]),"for l =",l)
        #Calculating the changes for the input layer beta and gammas if batch normalisation is being used.
        mb_grad_y = mb_delta
        if self.batch_normalisation == True:
            # print("grad_gamma = ",np.shape(grad_gamma[-l+2]),"for l =",l)
            # print("calculating grad_beta and grad_gamma for layer l=",l)
            # print("mb_grad_y shape =",np.shape(mb_grad_y))
            reshaped_mb_grad_y = np.reshape(mb_grad_y, (m, layer_size))
            grad_beta[0] = np.sum(reshaped_mb_grad_y.T, 1, keepdims=True)
            # print("normalised_mb_zs[-l+1] =",np.shape(normalised_mb_zs[0]))
            reshaped_normalised_mb_z = np.reshape(normalised_mb_zs[0], (m, layer_size))
            # print("reshaped_mb_grad_y =",np.shape(reshaped_mb_grad_y))
            grad_gamma[0] = np.sum((reshaped_mb_grad_y*reshaped_normalised_mb_z).T, 1, keepdims=True)
            # print("grad_gamma = ",np.shape(grad_gamma[-l+1]),"for l =",l)
        # print("finished")
        # for i in range(4):
            # print("i=",i)
            # print("grad_gamma = ",grad_gamma[i])
            # print("grad_beta = ",np.shape(grad_beta[i])
            # print("grad_w = ",np.shape(grad_w[i]))
        # print("grad_w[0] =",grad_w[0])
        # print("grad_w[3] =",grad_w[3])
        # print("grad_gamma =",grad_gamma)
        # print("grad_beta =",grad_beta)
        # print("grad_b =", grad_b)
        # return
        if self.batch_normalisation == False: return (grad_b, grad_w) #Finally we return a tuple of our cost gradient vectors wrt our biases and matrices wrt our weights for our whole network.
        if self.batch_normalisation == True: return (grad_gamma, grad_beta, grad_b, grad_w, layer_means, layer_variances) #If we are using batch normalisation then we instead return the cost gradient vectors wrt our weights, gammas and betas. 

    
    def evaluate(self, test_data):
        """Return the number of test inputs for which the neural
        ne twork outputs the correct result. Note that the neural
        network's output is assumed to be the index of whichever
        neuron in the final layer has the highest activation."""
        test_results = [(np.argmax(self.feedforward(x)), y)
                        for (x, y) in test_data]
        return sum(int(x == y) for (x, y) in test_results)

    def cost_derivative(self, output_activations, y):
        """Return the vector of partial derivatives \partial C_x /
        \partial a for the output activations."""
        return (output_activations-y)

    def display_weights(self, plot = True):
        for l in range(self.num_layers):
            layer = self.layers[l]
            if type(layer) == list:
                filter_size = layer[3]
                i = 1
                for w in self.weights[l]:
                    if plot == True:
                        filter_plot(w, i)
                    else:
                        w = np.reshape(w, (filter_size, filter_size))
                        print("Filter ",i," weights:")
                        print(w)
                        print("")
                    i = i + 1
        return
    
    def copy_network(self):
        network_copy = Network(self.layers, self.activation_function, self.padding)
        network_copy.weights = copy.deepcopy(self.weights)
        network_copy.biases = copy.deepcopy(self.biases)
        network_copy.experiences = copy.deepcopy(self.experiences)
        return network_copy
```

```python
#Miscellaneous functions
def sigmoid(z):
    """The sigmoid function."""
    return 1.0/(1.0+np.exp(-z))

def sigmoid_prime(z):
    """Derivative of the sigmoid function."""
    return sigmoid(z)*(1-sigmoid(z))

def relu(z):     #Defines our (shifted) rectified linear unit activation function.
    return np.array(np.where(z>0, z, 0),dtype=np.float32)
    
def relu_prime(z):    #Defines the derivative our rectified linear unit function.
    return np.array(np.where(z>0, 1, 0),dtype=np.float32)

def convolution_output_shape(input_layer_dimensions, filter_dimensions, padding):
    global convolution_time
    convolution_time_start = time.time()
    #Whether or not we use padding will effect the number of outputs of our convolution and thus the number of weights our convolutional layer needs. Note that with padding on we will require more weights.
    if input_layer_dimensions[0] < filter_dimensions[0]: (input_layer_dimensions, filter_dimensions) = (filter_dimensions, input_layer_dimensions)   #If the filter is larger than the layer then the role of the two are swapped in the convolution operation, hence we swap their roles here too.
    if padding == True: output_shape = [input_layer_dimensions[0]+filter_dimensions[0]-1, input_layer_dimensions[1]+filter_dimensions[1]-1] 
    if padding == False: output_shape = [input_layer_dimensions[0]-filter_dimensions[0]+1, input_layer_dimensions[1]-filter_dimensions[1]+1]
    convolution_time = convolution_time + time.time() - convolution_time_start
    return output_shape

#Function for determining the amount of padding needed in our (max) pooling process.
def pooling_padding_needed(input_layer_dimensions, pooling_filter_dimensions, stride):
    global pooling_time
    pooling_time_start = time.time()  
    padding_dimensions = [0,0]
    for i in [0, 1]:
        padding_dimensions[i] = stride*np.ceil((input_layer_dimensions[i]-pooling_filter_dimensions[i])/stride) - input_layer_dimensions[i] + pooling_filter_dimensions[i]
    #pooling_time = pooling_time + time.time() - pooling_time_start
    return padding_dimensions

#Function for performing max pooling on a given array.
def max_pooling(input_array, pooling_size, stride, return_indices = False):
    global pooling_time, index_time
    pooling_time_start = time.time()
    #If we have a stride which doesn't match our pooling size or if we want to keep track of each maximal index during our pooling step we can't simply use the block_reduce function so we need to max pool manually.
    if return_indices == True or stride != pooling_size:  
        input_array_dimensions = np.shape(input_array)[::-1]  #Finding the shape of the input array: note that np.shape returns the shape with the y-dimension first, hence we use [::-1] here to flip the order of the dimensions.
        #We then find the amount of padding needed to be added to our input array.
        padding_dimensions = pooling_padding_needed(input_array_dimensions, (pooling_size, pooling_size), stride)
        output_shape = (int((input_array_dimensions[1]+padding_dimensions[1]-pooling_size+stride)/stride), int((input_array_dimensions[0]+padding_dimensions[0]-pooling_size+stride)/stride))
        #Creating the padded array and inserting our input_array into the top left corner.
        padded_array = np.zeros((int(input_array_dimensions[1]+padding_dimensions[1]), int(input_array_dimensions[0]+padding_dimensions[0])))
        padded_array[0:input_array_dimensions[1], 0:input_array_dimensions[0]] = input_array
        #Perform max pooling process.
        output_array = np.zeros(output_shape)
        indices = []
        for i in range(output_shape[0]):  #Looping over y indices. 
            for j in range(output_shape[1]):   #Looping over x indices. 
                pooling_window = padded_array[i*stride:i*stride+pooling_size, j*stride:j*stride+pooling_size]
                index_time_start = time.time()
                max_index = pooling_window.argmax()
                row = max_index//pooling_size
                column = max_index%pooling_size
                max_value = pooling_window[row, column]
                output_array[i, j] = max_value
                if max_value == 0: indices.append(None)   #We don't append any values at padded indices (these will always be zero).
                else: indices.append([i*stride+row, j*stride+column])
                index_time = index_time + time.time() - index_time_start
        pooling_time = pooling_time + time.time() - pooling_time_start
        return output_array, indices
    else:  #In the special case that our stride matches our pooling size we can use the block_reduce function from skimage which typically runs faster.
        output_array = block_reduce(input_array, (pooling_size, pooling_size), np.max)
        return output_array
    
def get_similarity_gradient(filters):
    num_filters = len(filters)
    #First we pre-calculate some required values for our similarity derivative.
    inverse_norm_products = np.zeros((num_filters, num_filters))
    cosine_similarities = np.zeros((num_filters, num_filters))
    for i in range(num_filters):  #Since the calculated quantities are symmetric w.r.t. i and j we only need to populate the upper triangular entries of our matrix.
        for j in range(i+1, num_filters):  #We also don't populate the diagonal entries since we don't calculate similarities between any filter and itself.
            inverse_norm_products[i,j] = 1/(np.linalg.norm(filters[i])*np.linalg.norm(filters[j]))
            cosine_similarities[i,j] = np.dot(filters[i], filters[j])*inverse_norm_products[i,j]
    total_similarity = np.sum(cosine_similarities)
    #Using the symmetry w.r.t. i and j we can then populate the rest of our two matrices by adding the transpose of each of our current matrices.
    inverse_norm_products = inverse_norm_products + np.transpose(inverse_norm_products)
    cosine_similarities = cosine_similarities + np.transpose(cosine_similarities)
    sim_gradient = []
    for i in range(num_filters):
        sim_gradient.append(0)
        for filter, inverse_norm_product, similarity in zip(filters, inverse_norm_products[i], cosine_similarities[i]):
            sim_gradient[i] = sim_gradient[i] + filter*inverse_norm_product - filters[i]*similarity
    return sim_gradient

def get_similarity(filters):
    num_filters = len(filters)
    cosine_similarities = np.zeros((num_filters, num_filters))
    for i in range(num_filters):  #Since the cosine similarity are symmetric w.r.t. i and j we only need to populate the upper triangular entries of our matrix.
        for j in range(i+1, num_filters):  #We don't populate the diagonal entries since we don't calculate similarities between any filter and itself.
            cosine_similarities[i,j] = np.dot(filters[i], filters[j])/(np.linalg.norm(filters[i])*np.linalg.norm(filters[j]))
    return(np.sum(cosine_similarities))

def batch_normalise(mb_z, eps, gamma, beta, m): #Batch normalises the combined mini-batch activations of a given layer.
    mb_z = np.reshape(mb_z, (m, np.size(gamma))).T
    # print("mb_z during =",mb_z)
    means = np.mean(mb_z, 1, keepdims=True)
    # print("means during =",means)
    variances = np.var(mb_z, 1, keepdims=True)
    normalised_mb_z = (mb_z-means)/np.sqrt(variances+eps)
    normalised_mb_z = np.reshape(normalised_mb_z.T, (np.size(normalised_mb_z), 1))
    mb_gamma = np.tile(gamma, (m,1)) 
    mb_beta = np.tile(beta, (m,1))
    final_mb_z = mb_gamma*normalised_mb_z+mb_beta
    return final_mb_z, normalised_mb_z, means, variances  
```

```python
import random
import numpy as np

board, p1_score, p2_score, game_over, verbose, current_turn = None, 0, 0, False, False, 1

def input_move(move, show_game, game):  #This updates the board by inputting a nought (1) or cross (-1) into the chosen square, where the squares are arranged [0,1,2; 3,4,5; 6,7,8].
    global board, current_turn
    if game == "noughts and crosses":
        if move not in [0,1,2,3,4,5,6,7,8]:
            print("Invalid move; choose a number from 0 to 8")
            return
        if board[move] != 0:   #This checks to see whether the chosen square is full; if this if statement is true then the chosen move is invalid.
            print("Chosen square is full; choose a different square.")
            return
        if current_turn%2 == 1:
            board[move] = "1"
        else:
            board[move] = "-1"
        if show_game == True:   #Here we print the game out if the show_game flag is enabled.
            print("~~~~~~~~~~~~")
            print(np.reshape(board,(3,3)))
    if game == "connect 4":
        board = np.reshape(board, (6,7))
        for I in range(7):
            if move not in [0,1,2,3,4,5,6]:
                print("Invalid move; choose a number from 0 to 6")
                return
            if move == I:
                for i in range(5,-1,-1):
                    if board[i,I] == 0:
                        if current_turn%2 == 1:
                            board[i,I] = "1"
                            break
                        else:
                            board[i,I] = "-1"
                            break
                    elif board[0,I] != 0:
                        print("Chosen column is full; choose another column")
                        return
        if show_game == True:   #Here we print the game out if the show_game flag is enabled.
            print("~~~~~~~~~~~~")
            print(board)
    win_check(board, game)
    current_turn = current_turn + 1    #Incrementing the turn counter which keeps track of whose turn it is.
    return

def win_check(board, game):
    global current_turn
    if game == "noughts and crosses":
        if current_turn >= 5:       #Here we only check for a win after turn 4 since before this it isn't possible for either player to have three in a row.
            if board[0] == board[4] == board[8] != 0:    #Diagonal / win
                game_end(False, game)
                return
            if board[2] == board[4] == board[6] != 0:    #Diagonal \ win
                game_end(False, game)
                return
            for j in range(3):     
                if board[3*j] == board[3*j+1] == board[3*j+2] != 0: #Horizontal win
                    game_end(False, game)
                    return
            for j in range(3):     
                if board[j] == board[j+3] == board[j+6] != 0: #Vertical win
                    game_end(False, game)  
                    return
            if current_turn == 9:        #Draw
                game_end(True, game)
                return
    if game == "connect 4":
        board = np.reshape(board, (6,7))
        if current_turn >= 7:       #Here we only check for a win after turn 6 since before this it isn't possible for either player to have four in a row.
            for J in range(7):         #Vertical win
                for j in range(3):
                    if board[j,J] == board[j+1,J] == board[j+2,J] == board[j+3,J] != 0:
                        game_end(False, game)
                        return
            for J in range(6):         #Horizontal win
                for j in range(4):
                    if board[J,j] == board[J,j+1] == board[J,j+2] == board[J,j+3] != 0:
                        game_end(False, game)
                        return
            for J in range(3):         #Diagonal / win 
                for j in range(3,7):
                    if board[J,j] == board[J+1,j-1] == board[J+2,j-2] == board[J+3,j-3] != 0:
                        game_end(False, game)
                        return
            for J in range(3):         #Diagonal \ win
                for j in range(4):
                    if board[J,j] == board[J+1,j+1] == board[J+2,j+2] == board[J+3,j+3] != 0:
                        game_end(False, game)
                        return
            if current_turn == 42:        #Draw
                game_end(True, game)
    return

def game_end(draw, game = "noughts and crosses"):     #Here we take care of distribution the correct rewards to each player, updating the score and resetting the board for the next game.
    global board, p1_score, p2_score, game_over, current_turn, verbose
    if game == "noughts and crosses":
        board = np.reshape(board, (3,3))
    if game == "connect 4":
        board = np.reshape(board, (6,7))
    if draw == True:
        if verbose == True:
            print("Draw!")
            print("Final board:")
            print(board)
    else:
        if current_turn%2 == 1: 
            p1_score += 1
            if verbose == True:
                print("Player 1 wins!")
                print("Final board:")
                print(board)
        elif current_turn%2 == 0:
            p2_score += 1
            if verbose == True:
                print("Player 2 wins!")
                print("Final board:")
                print(board)
    game_over = True
    return

def get_valid_moves(board, game):
    valid_moves = []
    if game == "noughts and crosses":
        for i in range(9):
            if board[i] == 0:
                valid_moves.append(i)
    if game == "connect 4":
        board = np.reshape(board, (6,7))
        for i in range(7):
            if board[0,i] == 0:
                valid_moves.append(i)
    return(valid_moves)

def reset_game(game):  #Resets the board.
    global board, game_over, current_turn
    if game == "noughts and crosses":
        board = np.zeros((9),int)  #On the board '0' will be an empty space, '1' will be a nought and '-1' will be a cross.
    if game == "connect 4":
        board = np.zeros((6,7),int) #On the board '0' will be an empty space, '1' will be a red disc and '-1' will be a yellow disc.
    game_over = False
    current_turn = 1
    
def score():
    print("Player 1 score:", p1_score)
    print("Player 2 score:", p2_score)
```

```python
import time

def NeuralTD(total_games, lr, gamma, epsilon, p1 = None, p2 = None, decrease_epsilon = False, decrease_lr = False, target_network = False, game = "noughts and crosses", lmbda = 0, er_batch_size = 1, max_buffer_size = 100):
    global board, game_over, p1_score, p2_score, current_turn, update_time, standard_reward #Declaring some global variables used for gameplay and timings.
    start = time.time()    #Taking an initial start time used to time the entire runtime of the code.
    #Here we initialise our networks using (game specific) default architectures whenever we don't provide an input neural network for a particular player.
    #If we want to then we can enter p1 or p2 as pre-existing networks if we want to re-train them or specify a particular network architechture.
    #We can also choose to enter the string "random" in place of a network for either player to train the other network against a purely random opponent.
    if p1 == None:  
        if game == "noughts and crosses":
            p1 = Network([9, 40, 9])     
        if game == "connect 4":
            p1 = Network([42, 100, 7])
    if p2 == None:
        if game == "noughts and crosses":
            p2 = Network([9, 40, 9])     
        if game == "connect 4":
            p2 = Network([42, 100, 7])
    (p1_score, p2_score) = (0, 0) #Setting the initial scores to be 0-0.
    max_games = total_games  #Saving the number of games to be played overall (as total_games will be decremented later).
    (update_time, game_time, reshape_time, backprop_time, grad_time) = (0, 0, 0, 0, 0)  #For keeping track of how long certain parts of the code take to run.
    #Initialising arrays used to collect data on how many moves are made by p1 and p2 at the start and end of a given training session.
    (p1_moves_per_game_start, p1_moves_per_game_end, p1_invalid_move_attempts_per_game_start, p1_invalid_move_attempts_per_game_end) = ([], [], [], [])  
    (p2_moves_per_game_start, p2_moves_per_game_end, p2_invalid_move_attempts_per_game_start, p2_invalid_move_attempts_per_game_end) = ([], [], [], [])  
    #Here we save the values of the given epsilon and learning rate. note that these are used only if we set decrease_epsilon or decrease_lr to be true.
    initial_epsilon, initial_lr = epsilon, lr    
    standard_reward = 1   #Setting the standard reward/penalty given out during training after networks make winning/losing moves or invalid moves.
    
    #Here we loop over the total number of games to be played in our training session by playing a game to completion (with training) and decrementing total_games until it reaches 0 (at which point we stop training).
    while total_games > 0:   
        reset_game(game)  #Resetting the turn number and board for each new game.
        (current_player, waiting_player) = (p1, p2)  #Setting p1 to be the starting player and p2 to be the second player.
        (p1_moves_made, p2_moves_made, p1_invalid_move_attempts, p2_invalid_move_attempts) = (0, 0, 0, 0) #For keeping track of how many moves (and invalid moves) each player makes per game.
        #Here we loop over an individual game, during which we get our networks to choose moves and we update the networks according to the observed rewards and values of their chosen moves.
        while game_over == False: #As long as the game hasn't been ended by the previous player then the new current player chooses their move.
            initial_state = tuple(board.flatten())  #Storing and reshaping the initial board state into a vectorised form which is presentable to the current network player.
            chosen_action, action_values, invalid_move_made = select_move(current_player, initial_state, epsilon, lr, lmbda, game)  #Here we get our network to choose a move based on the current board.
            game_time_start = time.time()   #Timing the gameplay.
            initial_scores = (p1_score, p2_score)  #Recording the scores before the next move is made.
            input_move(chosen_action, False, game)    #Inputs the chosen action and updates the board.
            game_time = game_time + time.time() - game_time_start
            #Incrementing the number of moves made by the current player in this game.
            if current_player == p1: p1_moves_made = p1_moves_made + 1 
            elif current_player == p2: p2_moves_made = p2_moves_made + 1
            #Incrementing the number of invalid moves attempted by the current player in this game.
            if invalid_move_made == True and current_player == p1: p1_invalid_move_attempts = p1_invalid_move_attempts + 1 
            elif invalid_move_made == True and current_player == p2: p2_invalid_move_attempts = p2_invalid_move_attempts + 1
            #If the current player's move has ended the game then we need to update both player's most recent moves with an instantaneous reward/penalty respectively.
            if game_over == True: #Also note that since a player can't lose on their own turn, the game must have ended by the current player winning or by a draw.
                if (p1_score, p2_score) == initial_scores: reward = 0  #In the case of a draw (i.e. unchanged scores) we give both players a neutral 'reward' of 0.
                else: reward = standard_reward #Otherwise the current player must have won so we give them the standard reward (as specified above) and we penalise the waiting player by an equal amount.     
                if current_player != "random": save_experiences(current_player, initial_state, chosen_action, reward, None, game)   #Here give a positive reward for winning and we set next_state = None to signify that our next state is a terminal state, i.e. the game is over.
                if waiting_player != "random": save_experiences(waiting_player, wp_initial_state, wp_chosen_action, -reward, None, game)   #Here we give a negative reward for losing and set next_state = None to signify that our next state is a terminal state, i.e. the game is over.
            else:
                #If the game hasn't ended then the state created by the current player's move can now be used to update the values of the waiting player's chosen move (or specifically to update the waiting player's network).
                #We need this state since this is the next actionable state for the waiting player which is used to get the percieved value of this new state as required in the update step for the waiting player's network. 
                if current_turn != 2 and waiting_player != "random":    #We don't perform our network update if our network is the random player or if it's the second turn in the game since then the waiting player hasn't made any moves whose values need to be updated.
                    next_state = tuple(board.flatten())
                    save_experiences(waiting_player, wp_initial_state, wp_chosen_action, 0, next_state, game)   #Here we now save our next state since the game isn't over and also set our reward to 0 for the same reason.
            update_time_start = time.time()
            if current_player != "random":
                #if target_network == True and total_games%100 == 0 and current_turn <= 3:   #If we are training with a target network for our update steps then on the first turn of each 100th game for each player we sync up the target network with the actively updated network.
                #    current_player_target_network = current_player.copy_network()  
                if len(current_player.experiences) >= max_buffer_size:   #We perform experience replay every time we have filled out our experience buffer.
                    #if target_network == True: experience_replay(current_player, er_batch_size, lr, gamma, lmbda, game, current_player_target_network)    #Here we perform one update on our current player's network using a randomly assembled batch of 'experience tuples' taken from that player's experience buffer.
                    experience_replay(current_player, er_batch_size, lr, gamma, lmbda, game)
                    #After performing experience replay to update our network, we empty the experience buffer and start making new improved experiences with the improved network.
                    current_player.experiences = [] 
            update_time = update_time + time.time() - update_time_start
            (current_player, waiting_player) = (waiting_player, current_player) #Here we swap the active player with the waiting player.
            #Saving the most recent player's presented board state and their associated chosen action for use in our next update step.
            wp_initial_state = initial_state
            wp_chosen_action = chosen_action
            wp_action_values = np.copy(action_values)
            
        if total_games >= 0.9*max_games:   #Keeping track of the moves made and invalid moves made by each player in the first 10% of games (i.e. the start of training).
            p1_invalid_move_attempts_per_game_start.append(p1_invalid_move_attempts)
            p2_invalid_move_attempts_per_game_start.append(p2_invalid_move_attempts)
            p1_moves_per_game_start.append(p1_moves_made)
            p2_moves_per_game_start.append(p2_moves_made)
        if total_games <= 0.1*max_games:   #Keeping track of the moves made and invalid moves made by each player in the last 10% of games (i.e. the end of training).
            p1_invalid_move_attempts_per_game_end.append(p1_invalid_move_attempts)
            p2_invalid_move_attempts_per_game_end.append(p2_invalid_move_attempts)
            p1_moves_per_game_end.append(p1_moves_made)
            p2_moves_per_game_end.append(p2_moves_made)
            
        if total_games%(max_games*0.1) == 0: 
            print("~~~~~~~~~~~~~~~~~~~~~~~")
            print("Games left: ",total_games)    #Here we print out how many games are left at intervals of 10% of the total games to be played
            print("~~~~~~~~~~~~~~~~~~~~~~~")
            #If we want we can run the code and also linearly reduce our exploration rate "epsilon" or exponentially reduce our learning rate "lr" as training progresses
            if decrease_epsilon == True and total_games != max_games: epsilon = (total_games/max_games - 0.1)*initial_epsilon
            if decrease_lr == True and total_games != max_games: lr = 0.75*initial_lr
        total_games = total_games - 1  #Decrementing total_games after a game has been played to completion
    end = time.time()  #Taking a final time once all of our games have been played
    
    #Printing out the scores over all of the training session
    print("Final scores:")
    print("P1 score =",p1_score)
    print("P2 score =",p2_score)
    print("Draws = ",max_games-p1_score-p2_score)
    print("")
    #Printing out various timings during training
    print("Total time taken for",max_games,"games was {:.4f} seconds".format(end-start))
    print("Gameplay time = {:.4f}".format(game_time))
    print("Network update time = {:.4f}".format(update_time))
    print("Similar-board reshaping time = {:.4f}".format(reshape_time))
    print("")
    #Printing statistics about the moves made by each player at the start and end of training
    print("Average invalid moves by p1 per game at start = ",np.average(p1_invalid_move_attempts_per_game_start))
    print("Average invalid moves by p1 per game at end = ",np.average(p1_invalid_move_attempts_per_game_end))
    print("Average moves by p1 per game at start = ",np.average(p1_moves_per_game_start))
    print("Average moves by p1 per game at end = ",np.average(p1_moves_per_game_end))
    print("Average invalid moves by p2 per game at start = ",np.average(p2_invalid_move_attempts_per_game_start))
    print("Average invalid moves by p2 per game at end = ",np.average(p2_invalid_move_attempts_per_game_end))
    print("Average moves by p2 per game at start = ",np.average(p2_moves_per_game_start))
    print("Average moves by p2 per game at end = ",np.average(p2_moves_per_game_end))
    print("")
    return p1, p2   #Finally we return our training neural networks.
```

```python
def select_move(network, state, epsilon, lr, lmbda, game):   #Here we get the given neural network to select the move with the highest value in the current board state.
    global standard_reward, update_time
    state = np.reshape(state, (np.size(state), 1))
    invalid_move_made = False   #Setting a flag to determine whether our network tried to make an invalid move here.
    if network != "random": action_values = network.feedforward(state)   #Here we feed the current board state into our network and at returns the values it estimates each action to have in this state.
    if network == "random": action_values = None   #If we are playing against a random network then we don't need the action values as no update step is performed on our random opponent.
    valid_moves = get_valid_moves(state, game)                   #Creates a list of the indices for valid moves to be made, i.e. squares where our board is empty for noughts and crosses or where a column isn't full for connect 4.
    #Typically our agent will "exploit", i.e. chooses the greedy move, however here there is a small chance (epsilon) that our agent "explores", i.e. chooses a random move. Also if we entered "random" as our network then our move is always chosen randomly here.
    if (np.random.uniform() < epsilon or network == "random"):       
        chosen_action = random.choice(valid_moves) 
    else:
        chosen_action = np.argmax(action_values)    #As mentioned above, we usually "exploit" and choose the action with the highest value in our state.
        while chosen_action not in valid_moves:   #In the case that the network still thinks that the best move is an invalid move we need to punish this behaviour in the network and choose a valid move instead.
            invalid_move_made = True   #We keep track of the fact that our network tried to make an invalid move for the move statistics at the end of training.
            k = -random.randint(1,5000)   #Printing debug stats.
            backup_values = get_backup_values(network, state, chosen_action, -standard_reward, None, 0, game) #Here we punish the network for choosing an invalid move and force it to choose again.
            update_time_start = time.time()
            update_network(network, state, backup_values, lr, lmbda)   #Updating the network using a backup value of -1 in the position of the chosen invalid move.  
            update_time = update_time + time.time() - update_time_start  
            if k == 1:
                print("Invalid move chosen in position",chosen_action)
                print("Player's move values before and after and their difference:")
                after_values = network.feedforward(state)
                for (before, after) in zip(action_values, after_values):
                    print(before,"\t", after, "\t", after-before)
                print("")
            action_values_copy = np.copy(action_values)   #Creating a temporary copy of action values which we will use to ensure that a valid action is chosen next.
            for index in valid_moves:   #Having punished the network once we then ensure that it picks a valid move so that the game can continue.
                action_values_copy[index] = action_values_copy[index] + 100   #To ensure that the network now picks a valid move we temporarily increase only the valid move values by 100.
            chosen_action = np.argmax(action_values_copy)
    return chosen_action, action_values, invalid_move_made

def update_network(player, initial_state, backup_values, lr, lmbda = 0):
    initial_state = np.reshape(initial_state, (len(initial_state), 1))
    grad_b, grad_w = player.backprop(initial_state, backup_values)
    player.biases = [b-lr*gb for b, gb in zip(player.biases, grad_b)]
    player.weights = [w-lr*gw-lr*lmbda*w for w, gw in zip(player.weights, grad_w)]
    
def save_experiences(player, initial_state, chosen_action, reward, next_state, game):
    #Depending on our game we can speed up our learning by also training on geometrically different but still equivalent versions of our actual board. 
    #For example we can reflect our board horizontally to get a completely different state for our network to train on which is identical as far as gameplay is concerned.
    #In each case we save our equivalent boards alongside their equivalent actions, rewards, next states and next greedy actions in an 'equivalent experience' tuple saved to our experience buffer.    
    states = [initial_state]
    standard_actions = [[0,1,2],[3,4,5],[6,7,8]]  #Defining a matrix containing the usual position of each action which will be transformed below to a matrix with the equivalent positions of the actions after each transformation.
    current_experiences = [tuple([initial_state, chosen_action, reward, next_state])]    #Saving the original experience to a local experience buffer (they will all be saved to the overall buffer below).   
    if game == "noughts and crosses":
        reshaped_state = np.reshape(initial_state, (3,3))   #Reshaping our board from a vector into its natural shape so we can perform rotations/reflections on it.
        for k in range(1,4):
            equiv_initial_state = tuple(np.rot90(reshaped_state, k).flatten())   #Rotating our board by 90, 180 and 270 degrees gives an equivalent game state as far as noughts and crosses is concerned.
            if equiv_initial_state not in states:   #Here we make sure not to add any duplicate boards (i.e. rotationally symmetric boards) since we don't need to update the same exact board more than once.
                states.append(equiv_initial_state)
                #Creating a map from our old actions to their equivalent positions on our equivalent board - note that this requires our standard action board to be transformed via the inverse of the transformation applied to our actual board.
                remapped_actions = np.rot90(standard_actions,-k).flatten()   
                equiv_chosen_action = remapped_actions[chosen_action]    #Remapping the initial chosen action to its equivalent action.
                if next_state is not None: equiv_next_state = tuple(np.rot90(np.reshape(next_state, (3,3)), k).flatten())    #Reshaping the next state into its equivalent state.
                else: equiv_next_state = None
                current_experiences.append(tuple([equiv_initial_state, equiv_chosen_action, reward, equiv_next_state]))   #Saving the equivalent experience to our experience buffer.
        equiv_initial_state = tuple(np.fliplr(reshaped_state).flatten())   #Similarly we can reflect our board horizontally to get an equivalent game state.
        if equiv_initial_state not in states:
            states.append(equiv_initial_state)
            remapped_actions = np.fliplr(standard_actions).flatten()    #Creating a map from our old actions to their equivalent positions on our equivalent board.
            equiv_chosen_action = remapped_actions[chosen_action]    #Remapping the initial chosen action to its equivalent action.
            if next_state is not None: equiv_next_state = tuple(np.fliplr(np.reshape(next_state, (3,3))).flatten())    #Reshaping the next state into its equivalent state.
            else: equiv_next_state = None
            current_experiences.append(tuple([equiv_initial_state, equiv_chosen_action, reward, equiv_next_state]))   #Saving the equivalent experience to our experience buffer.
        equiv_initial_state = tuple(np.flipud(reshaped_state).flatten())   #Finally we can also reflect our board vertically to get an equivalent game state.
        if equiv_initial_state not in states:
            remapped_actions = np.flipud(standard_actions).flatten()    #Creating a map from our old actions to their equivalent positions on our equivalent board.
            equiv_chosen_action = remapped_actions[chosen_action]    #Remapping the initial chosen action to its equivalent action.
            if next_state is not None: equiv_next_state = tuple(np.flipud(np.reshape(next_state, (3,3))).flatten())    #Reshaping the next state into its equivalent state.
            else: equiv_next_state = None
            current_experiences.append(tuple([equiv_initial_state, equiv_chosen_action, reward, equiv_next_state]))   #Saving the equivalent experience to our experience buffer.
    if game == "connect 4":
        reshaped_board = np.reshape(initial_state, (6,7))   #Reshaping our board from a vector into its natural shape so we can perform rotations/reflections on it.
        equiv_initial_state = tuple(np.fliplr(reshaped_board).flatten())  #For connect 4 we can only get one extra state by flipping our board horizontally.
        if equiv_initial_state not in states:
            equiv_chosen_action = 6 - chosen_action   #Remapping the initial chosen action to its equivalent action.
            if next_state is not None: equiv_next_state = tuple(np.fliplr(np.reshape(next_state, (6,7))).flatten())    #Reshaping the next state into its equivalent state.
            else: equiv_next_state = None
            current_experiences.append(tuple([equiv_initial_state, equiv_chosen_action, reward, equiv_next_state]))   #Saving the equivalent experience to our experience buffer.   
    for experience in current_experiences: #Finally we save all of these experiences into the network's experience buffer.
        player.experiences.append(experience)    
    
#Here we determine the vector of backup values used in our backpropagation step where our chosen action has 'backup value' V_b given by our state/action value update step via V_new = V_old + alpha*(V_b - V_old). Our other values are typically unchangeed unless they exceed the expected range (-1, 1).
def get_backup_values(network, initial_state, chosen_action, reward, next_state, gamma, game):    #We set the default for next_action_value as none so that if game_over == True then we don't need to input next_action_value.
    global standard_reward
    initial_state = np.reshape(initial_state, (np.size(initial_state), 1))
    #The only update we potentially do to the unchosen move values is to set their target value to be in the range (-1, 1) if their current value slightly exceeds this limit (usually they will be within this range and this step won't do anything).
    initial_values = network.feedforward(initial_state)   #Determining the network's initial percieved values of each move in this state.
    initial_values = np.where(initial_values<-standard_reward, -standard_reward, initial_values)  #Ensuring that our target values don't go below -1.
    initial_values = np.where(initial_values>standard_reward, standard_reward, initial_values)  #Ensuring that our target values don't go above 1.
    if next_state == None: next_state_max_value = 0   #If the value of next_state is None then we set the value of the next state to 0 since it's a terminal state.
    else: 
        next_state = np.reshape(next_state, (np.size(next_state), 1))
        next_state_values = network.feedforward(next_state)
        next_state_valid_moves = get_valid_moves(next_state, game)
        for move in next_state_valid_moves: next_state_values[move] = next_state_values[move] + 10*standard_reward
        next_state_max_value = np.max(next_state_values) - 10*standard_reward  #Otherwise we feed the next state through our neural network and return the maximum value possible for each of the actions in this state.
    chosen_action_backup_value = reward + gamma*next_state_max_value      #For our chosen action we calculate the target value for our network using the TD(0) backup value equation: Vb = reward + gamma*(value of greedy action in our new state).
    backup_values = np.copy(initial_values)
    backup_values[chosen_action] = chosen_action_backup_value
    return backup_values

def experience_replay(network, batch_size, eta, gamma, lmbda, game, target_network = None):
    n = len(network.experiences)
    random.shuffle(network.experiences) #Shuffling our experiences to be in a random order.
    experience_batches = [network.experiences[k:k+batch_size] for k in range(0, n, batch_size)] #Partitioning our experiences into 'experience batches'.
    global_mean, global_variance = 0, 0 #Note that the global mean and variance are only needed if batch normalisation is being used.
    for experience_batch in experience_batches:
        mini_batch = [] #Using each experience we will generate training data in the form (initial_state, backup_values) which we store in this mini_batch array.
        for experience in experience_batch: #Loop over each experience in the current experience batch.
            initial_state, chosen_action, reward, next_state = experience   #Unpacking the experience tuple.
            initial_state = np.reshape(initial_state, (np.size(initial_state), 1))
            if target_network is not None: backup_values = get_backup_values(target_network, initial_state, chosen_action, reward, next_state, gamma, game)   #Using the experience to generate our backup values for each action.
            else: backup_values = get_backup_values(network, initial_state, chosen_action, reward, next_state, gamma, game)
            k = -random.randint(1,5000)   #Printing debug stats.
            if k == 1:  #Printing debug info.
                action_values_before = network.feedforward(initial_state)
            mini_batch.append((initial_state, backup_values)) #Here our initial state takes the place of our 'input' and the backup_values form our 'label'.
            #update_network(network, initial_state, backup_values, lr, lmbda)  #Backpropagating the error between the backup values and the current values through our network in order to update it.
        mb_means, mb_variances = network.update_mini_batch(mini_batch, eta, lmbda, n) #Using the mini batch we then train the network using our standard SGD routine.
        if network.batch_normalisation == True: 
            global_mean = np.add(global_mean, mb_means)
            global_variance = np.add(global_variance, mb_variances)
    #Calculating the overall global mean and variance from this set of experiences for use in network feedforward during actual gameplay.
    network.global_mean = global_mean/len(experience_batches)
    network.global_variance = batch_size/(batch_size-1)*global_variance/len(experience_batches)
            
```

```python
def networkPlayGames(total_games, player1 = "random", player2 = "random", show_game = False, game = "noughts and crosses"):
    global board, game_over, p1_score, p2_score, current_turn
    if player1 != "random":
        v1 = player1
    if player2 != "random":
        v2 = player2
    (p1_score, p2_score) = (0, 0)
    max_games = total_games
    start = time.time()
    while total_games > 0:
        reset_game(game)
        while game_over == False: #As long as the game hasn't been ended by p2 then p1 chooses their move
            state = np.reshape(board, (np.size(board), 1))
            valid_moves = get_valid_moves(state, game) 
            if player1 != "random":
                initial_values = v1.feedforward(state)
                chosen_action = np.argmax(initial_values)
                while chosen_action not in valid_moves:
                    for index in valid_moves:
                        initial_values[index] = initial_values[index] + 10
                    chosen_action = np.argmax(initial_values)
            else:
                chosen_action = random.choice(valid_moves)
            input_move(chosen_action, show_game, game)     #Inputs the chosen action and updates the board
            
            if game_over == False:     #As long as the game hasn't been ended by p1 then p2 chooses their move
                state = np.reshape(board, (np.size(board), 1))
                valid_moves = get_valid_moves(state, game) 
                if player2 != "random":
                    initial_values = v2.feedforward(state)
                    chosen_action = np.argmax(initial_values)
                    while chosen_action not in valid_moves:
                        for index in valid_moves:
                            initial_values[index] = initial_values[index] + 10
                        chosen_action = np.argmax(initial_values)
                else:
                    chosen_action = random.choice(valid_moves)
                input_move(chosen_action, show_game, game)     #Inputs the chosen action and updates the board
        if total_games%(max_games*0.1) == 0:
            print("Games left: ",total_games)
        total_games = total_games-1
    end = time.time()
    print("Final scores:")
    print("P1 score =",p1_score)
    print("P2 score =",p2_score)
    print("Draws =",max_games-p1_score-p2_score)
    print("Total time taken for",max_games,"games was {:.4f} seconds".format(end-start))
    return
```

```python
def playGames(total_games, player1 = "random", player2 = "random", show_game = False, game = "noughts and crosses"):
    global board, game_over, p1_score, p2_score, current_turn
    if player1 != "random":
        v1 = player1
    if player2 != "random":
        v2 = player2
    (p1_score, p2_score) = (0, 0)
    if game == "noughts and crosses":
        empty_board = tuple(np.zeros(9))
    if game == "connect 4":
        empty_board = tuple(np.zeros(42))
    max_games = total_games
    start = time.time()
    while total_games > 0:
        reset_game(game)
        while game_over == False: 
            state = np.reshape(board, (np.size(board), 1))
            valid_moves = get_valid_moves(state, game)                 #Creates a list of the indices where our board is zero, i.e. where there's an empty space
            if player1 != "random":
                if state not in v1:
                    v1[state] = empty_board       #Initialising any unseen states to assign value 0 to all moves
                if (np.array(v1[state]) == 0).all():     #If this is an unseen state then we choose randomly
                    print("Unseen state for p1")
                    chosen_action = random.choice(valid_moves)   
                else:
                    valid_actions = list(v1[state])          #When choosing our greedy move we give the valid actions a temporary extra +2 value to ensure that no invalid moves are chosen
                    for index in valid_moves:
                        valid_actions[index] = valid_actions[index] + 2
                    chosen_action = np.argmax(valid_actions)     #Since we are no longer training we will always choose the "greedy" option, i.e. we no longer explore
            else:
                chosen_action = random.choice(valid_moves)
            input_move(chosen_action, show_game, game)    #Inputs the chosen action and updates the board
            if game_over == False:     #As long as the game hasn't been ended by p1 then p2 chooses their move
                state = np.reshape(board, (np.size(board), 1))
                valid_moves = get_valid_moves(state, game)               #Creates a list of the indices where our board is zero, i.e. where there's an empty space
                if player2 != "random":
                    if state not in v2:
                        v2[state] = empty_board       #Initialising any unseen states to assign value 0 to all moves
                    if (np.array(v2[state]) == 0).all():     #If this is an unseen state then we choose randomly
                        print("Unseen state for p2")
                        chosen_action = random.choice(valid_moves)   
                    else:
                        valid_actions = list(v2[state])         #When choosing our greedy move we give the valid actions a temporary extra +2 value to ensure that no invalid moves are chosen
                        for index in valid_moves:
                            valid_actions[index] = valid_actions[index] + 2
                        chosen_action = np.argmax(valid_actions)     #Since we are no longer training we will always choose the "greedy" option, i.e. we no longer explore
                else:
                    chosen_action = random.choice(valid_moves)
                input_move(chosen_action, show_game, game)    #Inputs the chosen action and updates the board
        if total_games%(max_games*0.1) == 0:
            print("Games left: ",total_games)
        total_games = total_games-1
    end = time.time()
    print("Final scores:")
    print("P1 score =",p1_score)
    print("P2 score =",p2_score)
    print("Draws =",max_games-p1_score-p2_score)
    print("Total time taken for",max_games,"games was {:.4f} seconds".format(end-start))
    return
```
